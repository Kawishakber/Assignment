```python
def electricity(monthly_units):
    charges = 0
    if monthly_units > 1 and monthly_units <= 100:
        charges = monthly_units * 9.43
    elif monthly_units > 101 and monthly_units <= 200:
        charges = monthly_units * 10.29
    elif monthly_units > 201 and monthly_units <= 300:
        charges = monthly_units * 12.55
    elif monthly_units > 301 and monthly_units <= 400:
        charges = monthly_units * 17.38
    elif monthly_units > 401 and monthly_units <= 500:
        charges = monthly_units * 18.84
    elif monthly_units > 501 and monthly_units <= 600:
        charges = monthly_units * 19.76
    elif monthly_units > 601 and monthly_units <= 700:
        charges = monthly_units * 20.40
    elif monthly_units > 700:
        charges = monthly_units * 23.87
    else:
        charges = -1

    return charges
    
def uniform_charges(monthly_units):
    charges = monthly_units * 0.5715
    return charges
    
def fuel_charges(monthly_units):
    charges = monthly_units * 3.0114
    return charges

def sales_taxes(KE_Charges):
    charges = KE_Charges * 0.17
    return charges

def advance_taxes(KE_Charges):
    charges = KE_Charges * 0.075
    return charges

def Excize_duty(KE_Charges):
    charges = KE_Charges * 0.015
    return charges
    
def tv_fees(tv):
    charges = tv * 35
    return charges

def bill_main(monthly_units,tv):
    electric_charges = electricity(monthly_units)
    print("\nElectric charges for",monthly_units,f"Units is {int(electric_charges)} PKR")
    
    uniform_charge = uniform_charges(monthly_units)
    print("Uniform  charges for",monthly_units,f"Units is {int(uniform_charge)} PKR")
    
    fuel_charge = fuel_charges(monthly_units)
    print("Fuel charges for",monthly_units,f"Units is {int(fuel_charge)} PKR")
    
    KE_Charges = electric_charges + uniform_charge + fuel_charge
    
    sales_tax = sales_taxes(KE_Charges)
    print("Sales Tax for",monthly_units,f"Units is {int(sales_tax)} PKR")
    
    if KE_Charges > 20000:
        advance_tax = advance_taxes(KE_Charges)
        print("Advance Tax for",monthly_units,f"Units is {advance_tax} PKR")
    
    duty = Excize_duty(KE_Charges)
    print("Excize Duty Charges for",monthly_units,f"Units is {int(duty)} PKR")
    
    tv_fee = tv_fees(tv)
    print("TV License Fee for",tv,f"TV is PKR",tv_fee)
    
    total = electric_charges + uniform_charge + fuel_charge + sales_tax + duty + tv_fee
    print(f"\nYour Total Electricity Bill is {int(total)} PKR")

monthly_units = int(input("Enter Monthly Units: "))
tv = int(input("Enter number of TV: "))
bill_main(monthly_units,tv)
```
